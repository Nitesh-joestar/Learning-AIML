```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
insurance=pd.read_csv("insurance.csv")
insurance
```

```python
cat_encoder=OneHotEncoder(drop="first")
attributes=['smoker','region','sex']
cat_columns=cat_encoder.fit_transform(insurance[attributes])
```

```python
cat_encoder.get_feature_names_out()
```

```python
insurance_categories=pd.DataFrame(cat_columns.toarray(),columns=cat_encoder.get_feature_names_out())
insurance_categories
```

```python
insurance_pre=insurance
drop_attributes=['smoker','region','sex']
insurance_pre=insurance_pre.drop(drop_attributes,axis=1)
insurance_pre
```

```python
insurance_pre=insurance_pre.join(insurance_categories)
insurance_pre
```

```python
y_insurance_pre=insurance_pre[['charges']]
charges_attribute=['charges']
insurance_pre=insurance_pre.drop(charges_attribute,axis=1)
insurance_pre
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(insurance_pre,y_insurance_pre,test_size=0.2,random_state=42)
```

```python
y_test
```

```python
y_train.skew()
```

```python
y_train=np.log(y_train['charges'])
y_train.skew()
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler=StandardScaler()
fit_attributes=['age','bmi','children']
X_train_std=std_scaler.fit_transform(X_train[fit_attributes])
```

```python
X_test_std=std_scaler.transform(X_test[fit_attributes])
X_test_std
```

```python
X_train_std=pd.DataFrame(X_train_std,columns=std_scaler.get_feature_names_out())
X_train_std=X_train_std.join(insurance_categories)
X_train_std
```

```python
X_test_std=pd.DataFrame(X_test_std,columns=std_scaler.get_feature_names_out())
X_test_std=X_test_std.join(insurance_categories)
X_test_std
```

```python
y_test.skew()
```

```python
y_test=np.log(y_test['charges'])
y_test.skew()
```

```python
y_test=pd.DataFrame(y_test)
y_test
```

```python
y_train=pd.DataFrame(y_train)
y_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

lin_reg=LinearRegression()
lin_rmses=-cross_val_score(lin_reg,X_train_std,y_train,scoring="neg_root_mean_squared_error",cv=5)
pd.Series(lin_rmses).describe()
```

```python
lin_reg.fit(X_train_std,y_train)
```

```python
insurance_prediction=np.expm1(lin_reg.predict(X_test_std))
y_test=np.expm1(y_test)
```

```python
from sklearn.metrics import root_mean_squared_error
lin_rmse=root_mean_squared_error(insurance_prediction,y_test)
lin_rmse
```

```python
exp_y_test=np.expm1(y_test[['charges']])
exp_y_test
```

```python
exp_y_train=np.expm1(y_train[['charges']])
exp_y_train
```

```python
exp_y_test=y_test
exp_y_test.skew()
```

```python
lin_reg=LinearRegression()
lin_rmses=-cross_val_score(lin_reg,X_train_std,exp_y_train,scoring="neg_root_mean_squared_error",cv=5)
pd.Series(lin_rmses).describe()
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg=DecisionTreeRegressor(random_state=42)
tree_reg.fit(X_train_std,exp_y_train)
```

```python
tree_rmse=-cross_val_score(tree_reg,X_train_std,exp_y_train,scoring="neg_root_mean_squared_error",cv=5)
pd.Series(tree_rmse).describe()
```
