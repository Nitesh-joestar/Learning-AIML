```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
insurance=pd.read_csv("insurance.csv")
insurance
```

```python
cat_encoder=OneHotEncoder(drop="first")
attributes=['smoker','region','sex']
cat_columns=cat_encoder.fit_transform(insurance[attributes])
```

```python
cat_encoder.get_feature_names_out()
```

```python
insurance_categories=pd.DataFrame(cat_columns.toarray(),columns=cat_encoder.get_feature_names_out())
insurance_categories
```

```python
insurance_transform=insurance
```

```python
cat_attributes=['smoker','region','sex']
insurance_transform=insurance_transform.drop(cat_attributes,axis=1)
insurance_transform
```

```python
insurance_transform.describe()
```

```python
insurance_transform.skew()
```

```python
insurance_transform['charges']=np.log1p(insurance_transform['charges'])
```

```python
insurance_transform.skew()
insurance_transform
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler=StandardScaler()
insurance_standard=std_scaler.fit_transform(insurance_transform)
```

```python
df_insurance_standard=pd.DataFrame(insurance_standard,columns=std_scaler.get_feature_names_out())
df_insurance_standard
```

```python
df_insurance_standard=df_insurance_standard.join(insurance_categories)
df_insurance_standard
```
