my understanding of a classifier is this, there is an axis from -1 to 1
-1 being the model is confident its negative and 1 being positive
but how do u decide where the cutoff is where the positive starts from
that is where the threshold comes in
anything right of threshold the model claims is positive and left is negativve
by this way, if the threshold is very right. only highly confident answers go through
whereas if the threshold is quite low even mediumly confident answers go through
it has to predict something right

```python
y_score=cross_val_predict(sgd_clf,X_train,y_train_5,cv=3,method='decision_function')
y_score
```

```python
precisions,recall,threshold=precision_recall_curve(y_train_5,y_score)
```

```python
plt.plot(recall, precisions, linewidth=2)
plt.grid(True)
```

```python
recall
```

```python
idx_threshold=(precisions>=0.90).argmax()
threshold_for_90_precision=threshold[idx_threshold]
threshold_for_90_precision
idx_threshold
```

we are searching the precisions for a value greater than 0.9 and noting its index using the argmax function
finding the threshold at that index

```python
y_train_pred_90_prec=(y_score>=threshold_for_90_precision)
#predictions are now made true if the threshold is greater than 90 precision
```

```python
precision_score(y_train_5,y_train_pred_90_prec,)
```

WHAt we are doing here is you cant set threshold normally so we program it, we can get the decision value of each function, get the precision recall and decision PLOT IT. you can set the precision and recall manually
by finding its index oof precision or recall. like here

```python
idx_threshold=(recall>=0.90).argmax()
threshold_for_90_recall=threshold[idx_threshold]
threshold_for_90_recall
```

```python
y_train_pred_90_recall=(y_score>threshold_for_90_recall)
```

```python
recall_score(y_train_pred_90_recall,y_train_5,)
```

```python
idx_threshold=(recall>=0.90).argmax()
threshold_for_90_recall=threshold[idx_threshold]
threshold_for_90_recall
```

```python
min_recall=0.4
min_prec=0.9
both_conditions=(precisions[:-1]>=min_prec)&(recall[:-1]>=min_recall)
first_true_index=both_conditions.argmax()
threshold_for_both=threshold[first_true_index]
threshold_for_both
first_true_index
```
