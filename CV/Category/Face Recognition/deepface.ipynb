```python
from deepface import DeepFace as df
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
def show(img):
    img1=cv.imread(img)
    cv.imshow("window",img1)
    cv.waitKey(0)
    cv.destroyAllWindows()
```

```python
def verify(img1path,img2path):
    result=df.verify(img1_path=img1path,img2_path=img2path,
                     model_name='ArcFace',
                     detector_backend='retinaface')
    print(result)
    show(img1path)
    show(img2path)
```

```python
image='Celeb/marilyn-monroe-young-6-706326333.jpg'
```

```python
show(image)
```

```python
verify(image,'Celeb/mm2.jpeg')
```

```python
embedding=df.represent(image,model_name='ArcFace')
print(embedding)
```
